**Logarithm**  
4.

Roles/Usecases:
1. Logarithm can be used to sum probabilities, which can lead to more stable computaions. For instance, some probabilities can be very small, so when you want to use that probability in multiplication, it can result in a much more smaller probability that can lead to arithmetic underflow. Logarithm can help avoid this when dealing with small probabilities by allowing probabilities to be added instead of multiplied using the multiplication rule of logarithm.
2. Taking the logarithm of a ratio can better represent their changes in different directions. For example, take the ratio of $\frac{1024}{2}$ and $\frac{2}{1024}$. One is a very large number and the other one is a very small number and it makes it hard to understand the changes. But if you take the $\log_{2}$ of each number, you would yield 9 and -9. With these numbers, you can see their displacement is the same but in differen directions. One is positive and one is negative.
3. If you have a variable with a power distribution, you can take the log of it to make it more symmetrical and less skewed.
4. Logorithm is used as the function for cross-entropy loss in classification models. Cross-entropy loss function measures the differences between the predicted probability distribution of the class and the true label distribution of the class. The logarithm in this case is used to transform the two probability distributions to more accurately calculate the loss. Source used in [5].
5. In a linear regression model, logarithm can be used to linearize the input and output variables if they are non-linear or normalize the skewed data to make them more evenly distributed. Source used in [5].

**Correlation**  
5.
Correlation between two variable x and y describes how strong their relationship is and how one influence the other. A strong correlation can be seen as a cluster of data points being close to one another. The slope or the direction the data points are going from left to right can be seen as how much one influence the other. Other values such as the pearson correlation coefficient and spearman correlation can be used to quantify the strength and direction of the relationship. Causation means one variable is dependent on the other variable. Correlation does not always imply causation but causation can lead to a strong correlation between the two variables. One is example is the relationship of ice cream purchased and swim suits purchased. They both have a strong positive correlation because in the summer, they are popular due to hot weather. However, this strong correlation does not indicate one causes the other. Buying ice cream certainly do not cause buying a swim suit.

(a) $\bar{x} = \frac{5 + 8 + ... + 12.5 + 11.5}{20} = 8.125$  
$\bar{y} = \frac{62 + 74 + ... + 88 + 86}{20} = 73.9$

(b) $\sigma_x = \sqrt{\frac{\sum_{i=1}^{20} (x_i - \bar{x})^2}{20}} = \sqrt{\frac{(5 - 8.125)^2 + (8 - 8.125)^2 + ... + (12.5 - 8.125)^2 + (11.5 - 8.125)^2}{20} } = 3.478$  
$\sigma_y = \sqrt{\frac{\sum_{i=1}^{20} (y_i - \bar{y})^2}{20}} = \sqrt{\frac{(62 - 73.9)^2 + (74 - 73.9)^2 + ... + (88 - 73.9)^2 + (86 - 73.9)^2}{20} } = 12.996$

(c) $Cov(x, y) = \frac{\sum_{i=1}^{20} (x_i - \bar{x})(y_i - \bar{y})}{n} = \frac{(5 - 8.125)(62 - 73.9) + ... + (11.5 - 8.125)(86 - 73.9)}{20} = 44.938$

(d) $r = \frac{44.938}{3.478 * 12.996} = 0.994$

The pearson correlation between Hours of Study per Week and Exam Score is 0.994 which means Hours of Study per Week has a strong positive linear relationship with Exam Score. As the student studies longer, their exam score tend to be higher as well.

8.

The two correlation measures are almost the same. The pearson correlation is 0.001 less than the Spearman Rank Correlation Coefficien. Both indicate a strong positive relationship bewteen Hours of Study per Week and Exam Score. On the plot shown below, the direction of the data points are generally going up from left to right and they are very close to each other, indicating a strong positive relationship.

```python
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr

x = [5,8,7,9,11,4.5,10,3,12,6,7,10,2,13,8,14,5.5,3.5,12.5,11.5]
y = [62,74,69,76,85,60,80,55,90,65,70,82,50,92,78,95,63,58,88,86]

plt.scatter(x,y)
plt.xlabel('Hours of Study per Week')
plt.ylabel('Exam Score')
plt.title('Hours of Study per Week vs. Exam Score')

r, _ = pearsonr(x,y)
r_s, _ = spearmanr(x,y)

plt.annotate(f"Pearson: {r:.3f}", (0.7, 0.05), xycoords='axes fraction', fontsize=10, color='blue')
plt.annotate(f"Spearman: {r_s:.3f}", (0.7, 0.10), xycoords='axes fraction', fontsize=10, color='green')

plt.show()
```
